```python
from dateutil.relativedelta import relativedelta
from datetime import datetime
import pandas as pd
```

```python
def assign_numplan(strplan):
    if strplan == "ANNUAL":
        return 1
    elif strplan == "MONTHLY":
        return 12
    elif strplan == "QUARTER":
        return 4
    elif strplan == "HALF_ANNUAL":
        return 2
```

```python
def date_diff(date1,date2):
  date1 = datetime.strptime(date1,"%d/%m/%Y")
  date2 = datetime.strptime(date2,"%d/%m/%Y")
  difference_date =  relativedelta(date1,date2)
  return difference_date.years

```

```python
def generate_installments (contract_list):

  main_list = []


  diff = int(date_diff(contract_list[2],contract_list[1]))
  plan = assign_numplan(contract_list[6])
  iterator = diff*plan

  if contract_list[6] == "ANNUAL":
    for i in range(iterator):
      temp_list = []
      date = datetime.strptime(contract_list[1],"%d/%m/%Y") + relativedelta(years=1*i)
      contract_list[4] = contract_list[4] if contract_list[4] != '' else 0
      installment_price = ( int(contract_list[3]) - int(contract_list[4])) / iterator
      temp_list.append(i+1)
      temp_list.append(date.date())
      temp_list.append(installment_price)
      main_list.append(temp_list)
    return main_list

  elif contract_list[6] == "QUARTER":
    for i in range(iterator):
      temp_list = []
      date = datetime.strptime(contract_list[1],"%d/%m/%Y") + relativedelta(months=3*i)
      contract_list[4] = contract_list[4] if contract_list[4] != '' else 0
      installment_price = ( int(contract_list[3]) - int(contract_list[4])) / iterator
      temp_list.append(i+1)
      temp_list.append(date.date())
      temp_list.append(installment_price)
      main_list.append(temp_list)
    return main_list

  elif contract_list[6] == "MONTHLY":
    for i in range(iterator):
      temp_list = []
      date = datetime.strptime(contract_list[1],"%d/%m/%Y") + relativedelta(months=1*i)
      contract_list[4] = contract_list[4] if contract_list[4] != '' else 0
      installment_price = ( int(contract_list[3]) - int(contract_list[4])) / iterator
      temp_list.append(i+1)
      temp_list.append(date.date())
      temp_list.append(installment_price)
      main_list.append(temp_list)
    return main_list

  elif contract_list[6] == "HALF_ANNUAL":
    for i in range(iterator):
      temp_list = []

      date = datetime.strptime(contract_list[1],"%d/%m/%Y") + relativedelta(months=6*i)
      contract_list[4] = contract_list[4] if contract_list[4] != '' else 0
      installment_price = ( int(contract_list[3]) - int(contract_list[4])) / iterator
      temp_list.append(i+1)
      temp_list.append(date.date())
      temp_list.append(installment_price)
      main_list.append(temp_list)
    return main_list


```

```python
import csv

companies_list = []
with open("/content/Contracts.csv","r") as file_cont:
  reader_pen = csv.reader(file_cont)
  reader_pen.__next__()
  for row in reader_pen:
    companies_list.append(row)
print(companies_list)
```

```python
for i in range(len(companies_list)):

  main_list = generate_installments(companies_list[i])
  file_name = f'{companies_list[i][5]}.csv'


  with open(file_name, mode="w", newline="") as file:
      writer = csv.writer(file)
      writer.writerow(["Installment Number", "Installment Date", "Installment Price"])
      writer.writerows(main_list)
```
